## Visualizing MotionRugs on FishData

```python
df = pd.read_csv("..//data//fishdatashort.csv")
data = df.drop(columns = ['frame', 'id']).to_numpy().reshape(2000, 151, -1)
img_speed = HilbertCurve.MotionRugs(data, 256, 1, 2000, 2000)
img_aceleration = HilbertCurve.MotionRugs(data, 256, 2, 2000, 2000)
```
